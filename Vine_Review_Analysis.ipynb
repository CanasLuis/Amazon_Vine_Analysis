```python
import pandas as pd

df =pd.read_csv('vine_analysis.csv')

print(df)
```

```python
#more than 20 votes
total_votes = df[df["total_votes"]>=20]
total_votes
```

```python
#The data is filtered to create a DataFrame or table where the percentage of helpful_votes is equal to or greater than 50% 


df['%']=(df['helpful_votes']/df['total_votes'])*100
df_per=df[df['%']>=50]
df_per
```

```python
#The data is filtered to create a DataFrame or table where there is a Vine review
vine_review = df_per[df_per["vine"]=="Y"]

vine_review
```

```python
#The data is filtered to create a DataFrame or table where there isn’t a Vine review (5 pt)
vine_review_no = df_per[df_per["vine"]=="N"]

vine_review_no
```

```python
#The total number of reviews, the number of 5-star reviews, and the percentage 5-star reviews are calculated for all Vine and non-Vine reviews
#total number of reviews
vine_review.count()
```

```python
#number of 5 star reviews
vine_review_five = vine_review[vine_review['star_rating']==5]
#vine_review_five('star_rating').value_counts()
print(vine_review_five['star_rating'].value_counts())

#vine_review_five
```

```python
#percentage of all 5 star reviews

vine_review_five
```

```python
#total number of reviews for non vine

vine_review_no.count()
```

```python
#number of 5 star reviews for no vine
vine_review_no_five = vine_review_no[vine_review_no['star_rating']==5]
#vine_review_five('star_rating').value_counts()
print(vine_review_no_five['star_rating'].value_counts())
```
